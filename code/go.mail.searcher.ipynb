```python
import time

from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
SCROLL_PAUSE_TIME = 0.01
```

```python
#find links in top 10 pages from go.mail.ru
def find_in_top10(query, existing_links, driver):
    
    driver.get("https://go.mail.ru/")
    finded_links = []
    search = driver.find_element_by_name('q')
    search.send_keys(query)
    search.submit()
    time.sleep(4)
    
    last_height = driver.execute_script("return document.body.scrollHeight")
    pos = 0 
    for  i in range(15):
        time.sleep(SCROLL_PAUSE_TIME + 0.2)
        #print(driver.find_elements_by_class_name('SnippetResultInfo-leftBlock'))
        links = [finded.find_element_by_tag_name('a').get_attribute("href")
         for finded in driver.find_elements_by_class_name('SnippetResultInfo-leftBlock')]
        
        for link in links:
            #print("before:", link)
            link = '/'.join(link.split('/')[2:])
            if (link.endswith('/')):
                link = link[:-1]
                
            #print("after", link)
            if link in existing_links:
                finded_links.append(link)
        
        while True:
            time.sleep(SCROLL_PAUSE_TIME + 2)
            
            pos += 100
            driver.execute_script("window.scrollTo(0, {});".format(pos))


            

            new_height = driver.execute_script("return document.body.scrollHeight")

            if new_height == last_height:
                try:
                    next_button =  driver.find_elements_by_class_name('PageNavigator-itemBlock')[-1]
                    next_button.click()
                finally:
                    break
            last_height = new_height

    #driver.quit()
    return finded_links
```

```python
#find_in_top20('как заварить чай')
def create_existing_links(urlfile):
    url2doc_id = {}

    with open(urlfile, 'r') as f:
        
        for line in f:

            doc_id, url = line.split('\t')
            url2doc_id[url.rstrip()] = int(doc_id)
            
    return url2doc_id
```

```python
existing_links = create_existing_links('urls.numerate.txt')
```

```python
driver = webdriver.Firefox(executable_path=r'/home/evgeniy/driver/geckodriver')
finded_links = {}
with open('queries.numerate.txt', 'r') as f:
    with open('finded.csv', 'w') as fw:
        for line in f:
            query_id, query = line.split('\t')
            print(query)
            try:
                links = find_in_top10(query, existing_links, driver)
                for link in links:
                    fw.write('{},{}\n'.format(query_id, existing_links[link]))
            except :
                continue

        
```
